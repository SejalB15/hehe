```python
import matplotlib.pyplot as plt
```

```python
# Step 2: Define the function and its derivative
def f(x):
    return (x + 3)**2

def df(x):
    return 2 * (x + 3)   # derivative of (x+3)^2
```

```python
# Step 3: Initialize parameters
x_old = 2          # starting point
learning_rate = 0.1
max_iterations = 100
tolerance = 0.0001  # stop if the change is very small
```

```python
# Step 4: Gradient Descent loop
x_values = []  # to store all x for plotting
y_values = []

for i in range(max_iterations):
    grad = df(x_old)
    x_new = x_old - learning_rate * grad
    
    x_values.append(x_new)
    y_values.append(f(x_new))
    
    # check for convergence
    if abs(x_new - x_old) < tolerance:
        break
    
    x_old = x_new
```

```python
# Step 5: Output the result
print("Local minimum occurs at x =", round(x_new, 4))
print("Minimum value of function =", round(f(x_new), 4))
print("Total iterations:", i + 1)
```

```python
# Step 6: Plot the function and steps
import numpy as np
x = np.linspace(-10, 5, 100)
y = f(x)

plt.plot(x, y, label='Function y=(x+3)^2')
plt.scatter(x_values, y_values, color='red', label='Gradient descent steps')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gradient Descent Visualization')
plt.legend()
plt.show()
```
